```python
import pandas as pd
import numpy as np

np.random.seed(1337)
size = 8

random_matrix = np.random.randint(1, 10, (size, size))

print(f"{random_matrix.size = }")


index = np.random.choice(random_matrix.size, 10, replace=False)
print(f"{index = }")

# type converts it to float so that we can change elements to None
random_matrix = random_matrix * 1.0

random_matrix.ravel()[index] = None

scores = pd.DataFrame(
    random_matrix,
    index=[f"Player {i}" for i in range(1, size + 1)],
    columns=[f"Round {i}" for i in range(1, size + 1)],
)

scores
```

```python
np.random.seed(22)

n = 8
random_matrix = np.random.randint(1,10, (n,n))
index = np.random.choice(random_matrix.size, 45, replace = False)
print(f"index {index}")
print(f"Flattens the matrix: {random_matrix.ravel()}")

random_matrix = random_matrix*1.0 # to convert it to float as np.nan is float
random_matrix.ravel()[index] = None # returns contiguous flattened array
# numeric containers always use NaN 
print(random_matrix)
```

```python
scores = pd.DataFrame(random_matrix, index=[
                      f"Player {i}" for i in range(1, n+1)], columns=[f"Round {i}" for i in range(1,n+1)])
scores
```
